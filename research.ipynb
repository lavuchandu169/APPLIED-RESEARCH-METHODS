```python
# Import required libraries
import joblib
import pandas as pd
import os

# Load pre-trained models (assuming the models are in the current directory)
bed_model = joblib.load("bed_model.pkl")
staff_model = joblib.load("staff_model.pkl")
equipment_model = joblib.load("equipment_model.pkl")

# Define the prediction function
def predict(model, input_data):
    df = pd.DataFrame([input_data])
    return model.predict(df)[0]

# Example input data (similar to user input)
input_data = {
    "patient_load": 1,
    "staff_available": 2,
    "equipment_in_use": 3
}

# Get predictions from each model
bed_prediction = predict(bed_model, input_data)
staff_prediction = predict(staff_model, input_data)
equipment_prediction = predict(equipment_model, input_data)

# Generate recommendation
recommendation = (
    f"Allocate {round(bed_prediction)} beds, "
    f"assign {round(staff_prediction)} staff, "
    f"and allocate {round(equipment_prediction)} equipment units."
)

# Print results
print(f"Bed Allocation: {round(bed_prediction, 2)}")
print(f"Staff Optimization: {round(staff_prediction, 2)}")
print(f"Equipment Utilization: {round(equipment_prediction, 2)}")
print(f"Integrated Recommendation: {recommendation}")
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import joblib

# Load the dataset
data = pd.read_csv("hospital_data.csv")

# Define features and targets
X = data[["patient_load", "staff_available", "equipment_in_use"]]
y_bed = data["bed_allocation"]
y_staff = data["staff_optimization"]
y_equipment = data["equipment_utilization"]

# Split the data into training and testing sets
X_train, X_test, y_bed_train, y_bed_test = train_test_split(X, y_bed, test_size=0.2, random_state=42)
_, _, y_staff_train, y_staff_test = train_test_split(X, y_staff, test_size=0.2, random_state=42)
_, _, y_equipment_train, y_equipment_test = train_test_split(X, y_equipment, test_size=0.2, random_state=42)

# Train the models if not already trained
bed_model = RandomForestRegressor(random_state=42)
bed_model.fit(X_train, y_bed_train)

staff_model = RandomForestRegressor(random_state=42)
staff_model.fit(X_train, y_staff_train)

equipment_model = RandomForestRegressor(random_state=42)
equipment_model.fit(X_train, y_equipment_train)

# Make predictions for each target
bed_predictions = bed_model.predict(X_test)
staff_predictions = staff_model.predict(X_test)
equipment_predictions = equipment_model.predict(X_test)

# Calculate MSE and R² for bed allocation
bed_mse = mean_squared_error(y_bed_test, bed_predictions)
bed_r2 = r2_score(y_bed_test, bed_predictions)

print(f"Bed Allocation - MSE: {bed_mse:.2f}, R²: {bed_r2:.2f}")

# Calculate MSE and R² for staff optimization
staff_mse = mean_squared_error(y_staff_test, staff_predictions)
staff_r2 = r2_score(y_staff_test, staff_predictions)

print(f"Staff Optimization - MSE: {staff_mse:.2f}, R²: {staff_r2:.2f}")

# Calculate MSE and R² for equipment utilization
equipment_mse = mean_squared_error(y_equipment_test, equipment_predictions)
equipment_r2 = r2_score(y_equipment_test, equipment_predictions)

print(f"Equipment Utilization - MSE: {equipment_mse:.2f}, R²: {equipment_r2:.2f}")
```
